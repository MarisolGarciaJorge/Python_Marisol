```python
"""Vamos a hacer lo mismo , es decir, la relaci√≥n entre la situaci√≥n econ√≥mica del pa√≠s en la fecha del contacto con el cliente y la contratatci√≥n del nuevo producto"""
"""Vamos a crear un dataframe con los datos que me interesan"""
datos_economicos_generales = datos_banco [["emp.var.rate", "cons.price.idx", "cons.conf.idx", "euribor3m", "nr.employed", "y", "date"]]
datos_economicos_generales
```

```python
"""Vamos a extraer el a√±o de la fecha y sacar la media por a√±o de las variables """
"""Primero tenemos que traducir los meses al ingl√©s para que los reconozca"""
# Diccionario de meses en espa√±ol a ingl√©s
meses_es_en = {
	"enero": "January",
	"febrero": "February",
	"marzo": "March",
	"abril": "April",
	"mayo": "May",
	"junio": "June",
	"julio": "July",
	"agosto": "August",
	"septiembre": "September",
	"octubre": "October",
	"noviembre": "November",
	"diciembre": "December"
}

import re

def traducir_fecha(fecha):
	if pd.isnull(fecha):
		return fecha
	# Si ya es datetime, no hacer nada
	if isinstance(fecha, pd.Timestamp):
		return fecha
	# Reemplaza el nombre del mes en espa√±ol por ingl√©s
	for mes_es, mes_en in meses_es_en.items():
		fecha = re.sub(rf"-{mes_es}-", f"-{mes_en}-", str(fecha), flags=re.IGNORECASE)
	return fecha

# Aplica la traducci√≥n y convierte a datetime
datos_economicos_generales["date"] = datos_economicos_generales["date"].apply(traducir_fecha)
datos_economicos_generales_1 =datos_economicos_generales["date"] = pd.to_datetime(datos_economicos_generales["date"], errors="coerce", dayfirst=True)
```

```python
datos_economicos_generales["trimestre"] = datos_economicos_generales_1.dt.to_period("Q")
datos_economicos_generales
```

```python
"""Ahora vamos a agrupar por trimestre, sacando la media para las variables emp.variate, cons.price...y viendo si existe una correlaci√≥n entre estas variables en el trimestre de interacci√≥n con el cliente y la contrataci√≥n de producto"""
```

```python
"""Primero hay que convertir los valores de las columnas num√©ricas a n√∫meros  para que pueda hacer la media (si tiene comas no las reconoce como float)"""
# Convert columns with commas as decimal separators to float
for col in ['cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']:
	datos_economicos_generales[col] = datos_economicos_generales[col].replace(',', '.', regex=True)
	datos_economicos_generales[col] = pd.to_numeric(datos_economicos_generales[col], errors='coerce')

"""Para hacer la media s√≥lo tiene en cuenta las columnas num√©ricas"""
# Now group by year and calculate the mean only for numeric columns
numeric_cols = datos_economicos_generales.select_dtypes(include='number').columns
datos_economicos_generales_por_trimestre = datos_economicos_generales.groupby("trimestre")[numeric_cols].mean()
datos_economicos_generales_por_trimestre
"""Para tener el trimestre como columna y no como √≠ndice"""

datos_economicos_generales_por_trimestre.reset_index ()
```
